# Exploratory Data Analysis 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("data/processed/cleaned_baywheels-tripdata.csv")
```

```python
# Trip duration distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['trip_duration_minutes'], bins=50, kde=True)
plt.title('Trip Duration Distribution (min)')
plt.xlabel('Duration (min)')
plt.ylabel('Count')
plt.xlim(0, 60)
plt.show()
```

```python
# Trips by hour
sns.countplot(x='hour', data=df)
plt.title('Trips by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Trip Count')
plt.show()
```

```python
# Member vs Casual
sns.countplot(x='member_casual', data=df)
plt.title('Rider Type Distribution')
plt.show()
```

```python
# Heatmap by day of week and hour
pivot = df.groupby(['day_of_week', 'hour']).size().unstack()
sns.heatmap(pivot, cmap="YlGnBu")
plt.title('Trips Heatmap by Day and Hour')
plt.show()
```
